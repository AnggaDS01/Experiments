```python
class Layer_Dense:
    def __init__(self, n_inputs, n_neurons):
        self.weights = 0.001 * np.random.randn(n_inputs, n_neurons)
        self.biases = np.zeros((1, n_neurons))
    
    def forward(self, inputs):
        self.output = np.dot(inputs, self.weights) + self.biases

inputs = np.array([
    [1, 2, 3, 2.5],
    [2., 5., -1., 2],
    [-1.5, 2.7, 3.3, -0.8]
])

dense1 = Layer_Dense(inputs.shape[-1], 3)
dense1.forward(inputs)
dense2 = Layer_Dense(dense1.output.shape[-1], 2)
dense2.forward(dense1.output)
dense3 = Layer_Dense(dense2.output.shape[-1], 3)
dense3.forward(dense2.output)

print("First Dense Layer:")
print(f"inputs ordo: {inputs.shape}; weights1 ordo: {dense1.weights.shape}; matriks product: {dense1.output.shape}")
print(f"inputs: {inputs}")
print(f"weights: {dense1.weights}")
print(f"output: {dense1.output}")

print("\nSecond Dense Layer:")
print(f"inputs ordo: {dense1.output.shape}; weights2 ordo: {dense2.weights.shape}; matriks product: {dense2.output.shape}")
print(f"inputs: {dense1.output}")
print(f"weights: {dense2.weights}")
print(f"output: {dense2.output}")

print("\nThird Dense Layer:")
print(f"inputs ordo: {dense2.output.shape}; weights2 ordo: {dense3.weights.shape}; matriks product: {dense3.output.shape}")
print(f"inputs: {dense2.output}")
print(f"weights: {dense3.weights}")
print(f"output: {dense3.output}")
```
