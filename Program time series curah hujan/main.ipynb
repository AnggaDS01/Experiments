```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter
a = 1  # Anda bisa mengganti nilai a sesuai kebutuhan

# Fungsi solusi
def y(t, a):
    return (1/a*2) * (1 - np.cosh(a*(t-1)))

# Waktu t dari 0 hingga 3
t = np.linspace(0, 3, 400)

# Menghitung nilai y untuk setiap t
y_values = y(t, a)

# Plotting
plt.plot(t, y_values, label='$y(t) = \\frac{1}{a^2} (1 - \\cosh(a(t-1)))$')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(1, color='red', linestyle='--', label='t = 1')
plt.xlabel('$t$')
plt.ylabel('$y(t)$')
plt.title('Grafik Solusi $y(t)$')
plt.legend()
plt.grid(True)
plt.show()
```
